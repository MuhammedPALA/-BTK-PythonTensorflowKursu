```python
#Herhangi bir sınıfta oluşturulan def değerinin o sınıfa eklenen şeylerde kullanabiliyoruz
class Kopek():

    yilCarpani = 7
    
    def __init__(self,yas = 5):
        self.yas = yas
        self.insanYasınaCervilmisAttribute = yas *7

    def insanYasiHesapla(self):
        return self.yas * self.yilCarpani
```

```python
benimKopek = Kopek(3)
```

```python
benimKopek.yas
```

```python
benimKopek.insanYasiHesapla()
```

```python
benimKopek.insanYasınaCervilmisAttribute
```

inheritance

```python
class Hayvan():
    def __init__(self):
        print("Hayvan sınıfı init Çağırıldı")
    def method1(self):
        print("Hayvan sınıfı method1 Çağırıldı")
    def method2(self):
        print("Hayvan sınıfı method2 Çağırıldı")
        
```

```python
benimHayvanim = Hayvan()
```

```python
benimHayvanim.method1()
```

```python
benimHayvanim.method2()
```

```python
#Kedi sınıfı artık hayvan sınıfının özelliklerini de kullanabiliyor
class Kedi(Hayvan):
    def __init__(self):
        Hayvan.__init__(self)
        print("kedi sınıfı init Çağırıldı")
    #kedi hayvanın yaptığı şeyleri yapabiliyor ama hayvanlar kedinin yaptığı her şeyi yapamıyor mesela miyavla methodu
    def miyavla(self):
        print("miyav")
    #override üzerine yazmak
    def method1(self):
        print("kedi sınıfındaki method1 çağırıldı")
```

```python
benimKedi = Kedi()
```

```python
benimKedi.method1()
```

```python
benimKedi.miyavla()
```

```python
digerHayvan = Hayvan()
```

```python
digerHayvan.method1()
```

```python
digerKedi = Kedi()
```

```python
digerKedi.method1()
```

Polymorphism

```python
class Elma():
    def __init__(self,isim):
        self.isim = isim

    def bilgiVer(self):
        return self.isim + " 100 kaloridir "
    
```

```python
class Muz():
    def __init__(self,isim):
        self.isim = isim
    def bilgiVer(self):
        return self.isim + " 150 kaloridir "
```

```python
elma = Elma("elma")
```

```python
elma.bilgiVer()
```

```python
muz = Muz("muz")
```

```python
muz.bilgiVer()
```

```python
meyveListesi = [elma,muz]
```

```python
for meyve in meyveListesi:
    print(meyve.bilgiVer())
```
