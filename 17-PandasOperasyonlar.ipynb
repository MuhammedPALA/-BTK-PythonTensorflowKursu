Groupby

```python
maasSozlugu = {"Departman" : ["Yazılım","Yazılım","Pazarlama","Pazarlama","Hukuk","Hukuk"],
               "Calısan Ismi" : ["Ahmet","Mehmet","Atil","Burak","Zeynep","Fatma"],
               "Maas" : [100,150,200,300,400,500,]
              }
```

```python
maasDataFrame = pd.DataFrame(maasSozlugu)
```

```python
maasDataFrame
```

```python
groupObject = maasDataFrame.groupby("Departman")
```

```python
groupObject.count()
```

```python
groupObject.mean(numeric_only=True)
```

```python
groupObject.min()
```

```python
groupObject.max()
```
