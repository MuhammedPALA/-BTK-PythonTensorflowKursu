#ContinueBreakPass

```python
benimListem = [5,10,15,20,25,30]
```

```python
#Döngüyü durdurmak için kullanılır örnekteki gibi şartlandırılabilir 
for numara in benimListem:
    if numara == 15:
        break
    print(numara)
```

```python
#Döngüde verilen değeri atlayıp devam etmeyi sağlar
for numara in benimListem:
    if numara == 15:
        continue
    print(numara)
```

```python
#Döngüye bir şart eklenecek hatırlamadın ama kodların geri kalanını çalıştırman gerek pass kullanabilirsin
for numara in benimListem:
    pass
```
