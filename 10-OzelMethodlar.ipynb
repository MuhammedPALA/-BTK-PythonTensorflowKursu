```python
class Meyve():

    def __init__(self,isim,kalori):
        self.isim = isim
        self.kalori = kalori

    def __str__(self):
        return f"{self.isim} şu kadar kaloriye sahiptir: {self.kalori} "

    def __len__(self):
        return self.kalori
```

```python
muz = Meyve("Muz",150)
```

```python
print(muz)
```

```python
benimListem = [1,2,3,"a",4.5]
```

```python
print(benimListem)
```

```python
len(benimListem)
```

```python
len(muz)
```
