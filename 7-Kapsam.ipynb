Local, Enclosing, Global, Built-In 

```python
#Yani Mantık normalde en son verilen değer kabul edilirken fonksiyon vb kuramlarda dıştan içe doğru kabul edilir istenen katmanda değer değişiyorsa o değer istenen katmanda yok ise bir üst katmandaki değer alınır 
benimAdim = "PALA"
#Global

def benimFonksiyon():
    benimAdim = "Mahmut"
    #Enclosing
    def icFonksiyon():
        benimAdim = "Ayşe"
        #Local
        print(benimAdim)
    icFonksiyon()
        
```

```python
benimFonksiyon()
```

```python
benimAdim
```

```python
y = 10 

def yeniFonksiyon(y):
    print(y)
    y = 5
    print(y)
    return y
```

```python
yeniFonksiyon(3)
```

```python
y = yeniFonksiyon(3)
```

```python
y
```

```python
#bir fonksiyonda değiştirilen değerin globalde de değişmesini istiyorsak bu kodu kullanmamız gerek
y = 10 

def ornekFonksiyon():
    global y 
    y = 7
    print(y)
```

```python
ornekFonksiyon()
```
