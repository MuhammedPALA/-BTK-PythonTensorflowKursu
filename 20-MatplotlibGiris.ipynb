```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
#show gibi methodları yazmadan çalıştırır arka plan etkisi düşük ama her ide de çalışmaybailir kullanırken bu gibi şeylere dikkat et
%matplotlib inline
```

```python
yasListesi = [10,20,30,35,40,50,60,70,75]
```

```python
kiloListesi = [20,70,75,80,85,70,65,60,60]
```

```python
npYasList = np.array(yasListesi)
npKiloList = np.array(kiloListesi)
```

```python
npKiloList
```

```python
npYasList
```

```python
plt.plot(npYasList,npKiloList,"g")
```

```python
plt.plot(npYasList,npKiloList,"g")
plt.title("Kilonun Yaşa Göre Değişimi")
plt.xlabel("Yas")
plt.ylabel("Kilo")
plt.show()
```

```python
numpyDizisi1 = np.linspace(0,10,20)
```

```python
numpyDizisi1
```

```python
numpyDizisi2 = numpyDizisi1 ** 3
```

```python
numpyDizisi2
```

```python
plt.plot(numpyDizisi1,numpyDizisi2,"g--")
plt.show()
```

```python
#birden başka liste yapacaksak alt alta yada yan yana kullanılır
plt.subplot(1,2,1)
plt.plot(numpyDizisi1,numpyDizisi2,"r*-")
plt.subplot(1,2,2)# 1 sıra 2 kolon olacak şuan 2. grafiği çiziyorum.
plt.plot(numpyDizisi2,numpyDizisi1,"g--")
plt.show()
```

```python
benimFigure = plt.figure()
figureAxes = benimFigure.add_axes([0.1,0.1,0.3,0.3])
figureAxes.plot(numpyDizisi1,numpyDizisi2,"g")
figureAxes.set_xlabel("X Ekseni")
figureAxes.set_ylabel("Y Label")
figureAxes.set_title("Grafik Başlığı")

plt.show()
```

```python
figure2 = plt.figure()

eksen1 = figure2.add_axes ([0.1,0.1,0.9,0.9])
eksen2 = figure2.add_axes ([0.2,0.5,0.3,0.3])

eksen1.plot (numpyDizisi1, numpyDizisi2,"g")
eksen1.set_xlabel("x Ekseni") 
eksen1.set_ylabel("Y Ekseni") 
eksen1.set_title ("Ana Grafik Başlık")

eksen2.plot (numpyDizisi2, numpyDizisi1,"b")
eksen2.set_xlabel( "x Ekseni küçük") 
eksen2.set_ylabel( "Y Ekseni küçük") 
eksen2.set_title( "Küçük Grafik Başlık")

plt.show()
```

```python
(benimFigure, benimEksenler) = plt.subplots(nrows=1, ncols=2)

for eksen in benimEksenler:
    eksen.plot(numpyDizisi1,numpyDizisi2,"g")
    

#Tablolar arası mesafeyi ayarlar
plt.tight_layout()

plt.show()
    
```

```python
yeniFigur = plt.figure(dpi=1000)#dpi kaydederken de verilebilir

yeniEksen = yeniFigur.add_axes([0.1,0.1,0.9,0.9])

yeniEksen.plot(numpyDizisi1,numpyDizisi1 ,label = "numpyDizisi")
yeniEksen.plot(numpyDizisi1,numpyDizisi1 **2 ,label = "numpyDizisi **2")
yeniEksen.legend(loc=0)

plt.show()
```

```python
#tablo kaydetmek için 
yeniFigur.savefig("benimfigur.png",dpi=1000)
```
