```python
def bolmeIslemi(numara):
    return numara / 2
```

```python
bolmeIslemi(20)
```

```python
benimListem= list(range(1,11))
benimListem
```

```python
yeniListe = []
for eleman in benimListem:
    yeniListe.append(bolmeIslemi(eleman))
```

```python
yeniListe
```

map

```python
list(map(bolmeIslemi,benimListem))
```

```python
#yapay zeka yardımıyla oluşturuldu dersten bağımsız bir deneme olarak
def kontrolFonksiyonu():  # Buraya parantez eklendi
    try:
        veri = input("Lütfen bir kelime yazın: ")

        # Eğer sadece harflerden oluşmuyorsa hata fırlat
        if not veri.isalpha():
            raise ValueError("Sadece harflerden oluşan giriş yapmalısınız.")

        print("Girdiğiniz kelime:", veri)
        return "a" in veri

    except Exception as e:
        print("Lütfen String bir değer girin:", e)
        return False  # Hata durumunda sonuç False olsun

# Test etme
sonuc = kontrolFonksiyonu()
print("Kelime 'a' içeriyor mu?", sonuc)
```

```python
def kontrolFonksiyonu0 (string): 
    return "a" in string
```

```python
stringListesi = ["atıl", "atlas", "zeynep", "mehmet", "ahmet"]
```

```python
 sonucListesi = list(map(kontrolFonksiyonu0,stringListesi))
```

```python
sonucListesi.count(False)
```

filter

```python
#Sadece true değerleri filtreler
list(filter(kontrolFonksiyonu0,stringListesi))
```
