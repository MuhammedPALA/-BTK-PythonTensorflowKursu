```python
#Verilen anahtar kelimeye değer veya değere kelime atanabilir  
bSozluk = {"anahtarkelime" : "deger"}
```

```python
type(bSozluk)
```

```python
bSozluk["anahtarkelime"]
```

```python
#Örnek sözlük oluşturalım
bKaloriSozlugum = {"elma" : 100, "karpuz" : 200, "muz" :300 }
```

```python
bKaloriSozlugum ["muz"]
```
