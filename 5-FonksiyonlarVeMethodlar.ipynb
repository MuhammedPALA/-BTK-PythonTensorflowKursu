args & kwargs

```python
#yapılan input işlemlerde her bir değere değişken atamadan sadece fonksiyonu vererek istediğimiz kadar değer(argüman) girilebilir
def yeniToplama(*args):
    return sum(args)
```

```python
yeniToplama(10,20,30,40)
```

```python
yeniToplama(50,60)
```

```python
#return etmeden tuple döndürüyor o yüzden fonksiyona return eklemyi unutmayalım
def benimFonksiyonum(*args):
    # print(args)
    return args
```

```python
type(benimFonksiyonum(20,30,40))
```

```python
def ornekFonksiyon(**kwargs):
    #print(args)
    return (kwargs)
```

```python
type(ornekFonksiyon(muz = 100, elma = 200, ananas = 300))
```

```python
def keyWordKontrol(**kwargs):
    if "PALA" in kwargs:
        print("PALA var")
    else: 
        print("PALA yok")
```

```python
keyWordKontrol(PALA = 83, MERT = 72, ALİ = 68) 
```

```python
keyWordKontrol(TUNA = 83, MERT = 72, ALİ = 68) 
```
