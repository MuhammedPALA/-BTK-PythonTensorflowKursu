#enumerate

```python
#verilen eleman kümesini indeksliyor yani sql deki count sütunu gibi ve bu işlemin çıktıları tuple oluyor 
for eleman in enumerate(list(range(5,15))):
    print(eleman)
```

```python
#istersek çıktıya değer atayıp sadece indeksi yada verilen değerleri alabiliriz
for (indeks,numara) in enumerate(list(range(5,15))):
    #print(indeks)
    print(numara)
    
    
```
