```python
import numpy as np
import pandas as pd
```

```python
data1 = {
"Indeks": [0, 1, 2, 3],
"Isim": ["Ahmet", "Mehmet", "Zeynep", "Atil"],
"Spor": ["Koşu", "Yüzme", "Koşu", "Basketbol"],
"Kalori": [100, 200, 300, 400]
}
df1 = pd.DataFrame(data1)

data2 = {
"Indeks": [4, 5, 6, 7],
"Isim": ["Osman", "Levent", "Atlas", "Fatma"],
"Spor": ["Koşu", "Yüzme", "Koşu", "Basketbol"],
"Kalori": [200, 100, 50, 300]
}
df2 = pd.DataFrame(data2)

data3 = {
"Indeks": [8, 9, 10, 11],
"Isim": ["Ayse", "Mahmut", "Duygu", "Nur"],
"Spor": ["Koşu", "Yüzme", "Badminton", "Tenis"],
"Kalori": [300, 400, 500, 250]
}
df3 = pd.DataFrame(data3)
```

Concatenation

```python
#dataframe birleştirmek için kullanılır varsayılan axis 0 istersek 1 olarak değiştirebiliriz 
pd.concat([df1,df2,df3])
```

Merge

```python
mergeSozluk1 = {"Isim" : ["Ahmet", "Mehmet", "Zeynep", "Atıl"],
                "Spor" : ["Kosu", "Yüzme", "Kosu", "Basketbol"],
               }
```

```python
mergeDataFrame1 = pd.DataFrame(mergeSozluk1)
```

```python
mergeDataFrame1
```

```python
mergeSozluk2 = {"Isim" : ["Ahmet", "Mehmet", "Zeynep", "Atıl"],
                "Kalori" : [100,200,150,250],
               }
```

```python
mergeDataFrame2 = pd.DataFrame(mergeSozluk2)
```

```python
mergeDataFrame2
```

```python
#merge dataframelerde birbirine uyumlu olması kaydıyla stünları birleştirir referans sütun on=seçilensütun olarak girilir
pd.merge(mergeDataFrame1,mergeDataFrame2,on="Isim")
```

```python
maasSozluk = {"Isim" : ["Atil","Zeynep","Mehmet","Ahmet"],
             "Departman" : ["Yazılım","Satıs","Pazarlama","Yazılım"],
             "Maas" : [200,300,400,500]
             }
```

```python
maasDataFrame = pd.DataFrame(maasSozluk)
```

```python
maasDataFrame
```

```python
#farklı olanları sıralar aynı olanları tekrar getirmez
maasDataFrame["Departman"].unique()
```

```python
#kaç departman var
maasDataFrame["Departman"].nunique()
```

```python
#Hangi departmanda kaç kişi olduğunu getirir
maasDataFrame["Departman"].value_counts()
```

```python
def bruttenNete(maas):
   return maas * 0.27 + maas
```

```python
#apply uygula demek bu örnekte maasDataFrame den maas sütnunu getir ve bruttenNete fonksiyonunu uygula diyoruz
maasDataFrame["Maas"].apply(bruttenNete)
```
